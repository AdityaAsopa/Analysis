```python
#this python program is to generate a number of trials of 1 seconds long poisson train with an average rate of 30 Hz
import numpy as np
import matplotlib.pyplot as plt
import random
import math

#generate a poisson train with an average rate of 30 Hz
def poisson_train(rate, duration=1):
    t = 0
    spike_train = []
    while t < duration:
        t += random.expovariate(rate)
        if t < duration:
            spike_train.append(t)
    return spike_train

#generate n trials of poisson train
def poisson_trials(rate, time=1, num_trials=5):
    trials = []
    for i in range(num_trials):
        trials.append(poisson_train(rate, duration=time))
    return trials
```

```python
def plot_spike_train(spike_train_trials):
    for i, spike_train in enumerate(spike_train_trials):
        plt.eventplot(spike_train, lineoffsets=i+1,)
    plt.xlabel('Time (s)')
    plt.ylabel('Trial')
    plt.title('Poisson Spike Train')
    plt.show()
```

```python
# plot the spike train
plt.figure()
spike_train_trials = poisson_trials(30, time=1, num_trials=5)
plot_spike_train(spike_train_trials)
# save as svg
plt.savefig(r'C:\Users\adity\OneDrive\NCBS\Lab\Thesis\Images\poisson_spike_train.svg')
plt.savefig(r'C:\Users\adity\OneDrive\NCBS\Lab\Thesis\Images\poisson_spike_train.png')
```
